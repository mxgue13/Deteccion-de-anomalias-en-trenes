# **1. Decision Tree Classifier**

```python
#@title Tunnig de parametros (max_depth)
def prueba(est):
    from sklearn.tree import DecisionTreeClassifier
    means, stds = [], [] # Listas para almacenar las medias y desviaciones estndar de las precisiones
    max_depth_range = range(1,30,1) # Rango de valores para el parmetro max_depth
    # Iteramos sobre los valores de max_depth
    for max_depth_c in max_depth_range:
        #print (max_depth,)
        est = DecisionTreeClassifier(max_depth=max_depth_c,random_state=21)
        est.fit(X_train,y_train)
        s=accuracy_score(est.predict(X_test), y_test)
        means.append(np.mean(s))
        stds.append(np.std(s))

    means = np.r_[means]
    stds  = np.r_[stds]

    plt.plot(max_depth_range, means, label="mean", color="black")
    # Rellenamos el rea entre las curvas de mean-std y mean+std para visualizar la variabilidad de las precisiones
    plt.fill_between(max_depth_range, means-stds, means+stds, color="blue", alpha=.5, label="std")
    plt.xlabel("max depth")
    plt.ylabel("accuracy")
    plt.legend()
```

```python
#@title Grafica
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

prueba(DecisionTreeClassifier())
```

```python
#@title Tunnig de parametros (criterion)
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.utils import shuffle
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

def criterion1(X, y):

    # Estimadores con sus respectivos hiperparmetros
    est1 = DecisionTreeClassifier(criterion='gini', max_depth=6, random_state=21)
    est2 = DecisionTreeClassifier(criterion='entropy', max_depth=6, random_state=21)
    est3 = DecisionTreeClassifier(criterion='log_loss', max_depth=6, random_state=21)

    # Entrenar los estimadores
    est1.fit(X_train, y_train)
    est2.fit(X_train, y_train)
    est3.fit(X_train, y_train)


    # Calcular la exactitud en el conjunto de validacin
    score1 = accuracy_score(est1.predict(X_test), y_test)
    score2 = accuracy_score(est2.predict(X_test), y_test)
    score3 = accuracy_score(est3.predict(X_test), y_test)

    return score1, score2, score3

X, y = load_iris(return_X_y=True)

score1, score2, score3 = criterion1(X, y)

f = plt.figure(figsize=(12,3))
f.subplots_adjust(wspace=0.5)

ax1 = f.add_subplot(121)
ax1.bar([0,1,2], [score1,score2,score3], capsize=10)
ax1.set_xticks(range(0,3))
ax1.set_xticklabels(('gini', 'entropy', 'log_loss'))
ax1.set_ylabel('Accuracy')
ax1.text(-0.33, 1.01, str(score1)[:8])
ax1.text(0.68, 1.01, str(score2)[:8])
ax1.text(1.8, 1.01, str(score3)[:8])
ax1.axis(ymin=0, ymax=1)  # Ajuste del rango del eje y
plt.show()
```

```python
#@title Tunnig de parametros (kernel)
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.utils import shuffle
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

def kernel(X, y):

    # Estimadores con sus respectivos hiperparmetros
    est2 = SVC(kernel="poly",degree=15,C=15,random_state=21)
    est3 = SVC(kernel= "rbf",C=15,gamma=5,random_state=21)

    # Entrenar los estimadores
    est2.fit(X_train, y_train)
    est3.fit(X_train, y_train)


    # Calcular la exactitud en el conjunto de validacin
    score2 = accuracy_score(est2.predict(X_test), y_test)
    score3 = accuracy_score(est3.predict(X_test), y_test)

    return score2, score3

X, y = load_iris(return_X_y=True)

score2, score3 = kernel(X, y)

f = plt.figure(figsize=(12,3))
f.subplots_adjust(wspace=0.5)

ax1 = f.add_subplot(121)
ax1.bar([0,1], [score2,score3], capsize=10)
ax1.set_xticks(range(0,2))
ax1.set_xticklabels(('poly', 'rbf'))
ax1.set_ylabel('Accuracy')
ax1.text(-0.33, 1.01, str(score2)[:8])
ax1.text(0.68, 1.01, str(score3)[:8])
ax1.axis(ymin=0, ymax=1)  # Ajuste del rango del eje y
plt.show()
```

# Comparacin de metodos

```python
#@title Comparacion de metodos para escoger el mejor
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.utils import shuffle
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

def mejor(X, y):

    # Estimadores con sus respectivos hiperparmetros
    est1 = DecisionTreeClassifier(criterion='gini', max_depth=7, random_state=21)
    est2 = RandomForestClassifier(criterion="gini", n_estimators=58,max_depth=13,random_state=21)
    est3 = SVC(kernel="rbf",C=15,gamma = 5,random_state=21)

    # Entrenar los estimadores
    est1.fit(X_train, y_train)
    est2.fit(X_train, y_train)
    est3.fit(X_train, y_train)


    # Calcular la exactitud en el conjunto de validacin
    score1 = accuracy_score(est1.predict(X_test), y_test)
    score2 = accuracy_score(est2.predict(X_test), y_test)
    score3 = accuracy_score(est3.predict(X_test), y_test)

    return score1, score2, score3

X, y = load_iris(return_X_y=True)

score1, score2, score3 = mejor(X, y)

f = plt.figure(figsize=(12,3))
f.subplots_adjust(wspace=0.5)

ax1 = f.add_subplot(121)
ax1.bar([0,1,2], [score1,score2,score3], capsize=10)
ax1.set_xticks(range(0,3))
ax1.set_xticklabels(('DTC', 'RFC', 'SVC'))
ax1.set_ylabel('Accuracy')
ax1.text(-0.33, 1.01, str(score1)[:8])
ax1.text(0.68, 1.01, str(score2)[:8])
ax1.text(1.8, 1.01, str(score3)[:8])
ax1.axis(ymin=0, ymax=1)  # Ajuste del rango del eje y
plt.show()
```
